- **Physical geometric parameters**  
  Length, ramp angles, bumper curvature, roof curvature, panel slopes, hood angle, etc.

  Drag coefficient (Cd), lift/downforce (Cl), and other flow-derived metrics.

  Approximate cabin volume, frontal area, interior shape metrics.

Your model must learn correlations between them to generate valid completions.

To simulate real engineering constraints, **some features are revealed** (the known physics/performance requirements) and others are **masked**.  
Your AI Copilot must generate **many plausible completions** for these masked (free) parameters.

### Data Preprocessing and Missing Value Handling

This is a critical section where we prepare our data for the VAE model. Key points:

**Missing Value Conventions:**
- In CSV files: `-1` indicates missing values
- In mask files: `True` = missing, `False` = observed
- For PyTorch models: We convert to `1` = observed, `0` = missing (standard convention)

**Why This Matters:**
Our VAE needs to distinguish between observed values (which provide constraints) and missing values (which need to be generated). The mask tells the model which values to trust and which to predict.

```python
# Correcting skeweed distributions
class FeatureTransformer:
    """
    Handles preprocessing for car design features, specifically applying the 
    Yeo-Johnson transformation to skewed features (28-34) to make them more Gaussian.
    
    The transformer is fitted only on non-missing data to ensure robustness.
    """
    def __init__(self, feature_indices_to_transform: List[int]):
        """
        Initializes the transformer.
        :param feature_indices_to_transform: A list of 0-indexed column indices 
                                             corresponding to the skewed features.
        """
        self.indices = feature_indices_to_transform
        # Use a list of transformers, one for each feature
        self.transformers: List[PowerTransformer] = [
            PowerTransformer(method='yeo-johnson', standardize=True) 
            for _ in feature_indices_to_transform
        ]
        print(f"Transformer initialized for {len(self.indices)} features.")

    def fit(self, X_train_NaN: pd.DataFrame):
        """
        Fits the PowerTransformer to the non-missing training data.
        
        :param X_train_NaN: The training data as a DataFrame, with missing values 
                            represented by NaN.
        """
        if not self.indices:
            print("No features configured for transformation.")
            return

        print(f"Fitting PowerTransformer on non-missing data for indices: {self.indices}...")
        
        # 1. Select the columns that need transformation
        cols_to_transform = X_train_NaN.iloc[:, self.indices]
        
        # 2. Iterate through each feature and fit its dedicated transformer
        for i, col_index in enumerate(self.indices):
            feature_data = cols_to_transform.iloc[:, i].dropna().values.reshape(-1, 1)
            
            if len(feature_data) > 0:
                self.transformers[i].fit(feature_data)
                
        print("✓ PowerTransformer fitted successfully.")

    def transform(self, X_NaN: Union[pd.DataFrame, np.ndarray]) -> np.ndarray:
        """
        Applies the fitted transformation to the input data.
        
        :param X_NaN: The data (train, validation, or test) with NaN for missing values.
        :return: A NumPy array with the specified features transformed.
        """
        if not self.indices:
            return X_NaN.values if isinstance(X_NaN, pd.DataFrame) else X_NaN
            
        X_out = X_NaN.values.copy() if isinstance(X_NaN, pd.DataFrame) else X_NaN.copy()
        
        print(f"Transforming data for indices: {self.indices}...")
        
        for i, col_index in enumerate(self.indices):
            # Extract the column data (including NaNs)
            col_data = X_out[:, col_index].reshape(-1, 1)
            
            # Identify where the NaNs are
            nan_mask = np.isnan(col_data).flatten()
            
            # Only transform non-NaN values
            if np.sum(~nan_mask) > 0:
                # Apply the transformation
                transformed_values = self.transformers[i].transform(col_data[~nan_mask])
                
                # Place transformed values back into the output array
                X_out[~nan_mask, col_index] = transformed_values.flatten()

        return X_out
    
```

```python
# Features 28-34 correspond to 0-indexed positions 27 through 33
SKEWED_INDICES = list(range(27, 34)) 

X_train_original_c = X_train_original.copy()

df_X_train_original_c = pd.DataFrame(X_train_original_c)

# 2. Initialize and fit the transformer
preprocessor = FeatureTransformer(SKEWED_INDICES)
preprocessor.fit(df_X_train_original_c)
    
# 3. Apply the transformation
transformed_data_np = preprocessor.transform(df_X_train_original_c)
transformed_df = pd.DataFrame(transformed_data_np)
    
# 4. Check the results (Should be close to zero mean, close to zero median)
print("\n--- Transformed Skewness Check (Feature 28, index 27) ---")
print(f"Transformed Mean: {transformed_df.iloc[:, 27].mean():.3f}")
print(f"Transformed Median: {transformed_df.iloc[:, 27].median():.3f}")
    
# Note: NaN values remain NaN after transformation
print(f"NaN count after transform: {np.isnan(transformed_data_np[:, 27]).sum()}")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
import numpy as np
import pandas as pd
from sklearn.preprocessing import QuantileTransformer, MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import math

# Import your utils
from utils import load_dataset_splits

# ==========================================
# 1. GRID PREPROCESSOR (The Fix)
# ==========================================
class GridPreprocessor:
    def __init__(self, discrete_threshold=25):
        self.threshold = discrete_threshold
        self.discrete_cols = []
        self.continuous_cols = []
        self.mappings = {} # {col_idx: {val: grid_val}}
        self.reverse_mappings = {} # {col_idx: {grid_val: val}}
        self.q_scaler = QuantileTransformer(output_distribution='normal', n_quantiles=2000)
        
    def fit(self, data):
        n_features = data.shape[1]
        
        # 1. Detect Types
        print("[GridPreprocessor] Analyzing Feature Types...")
        for i in range(n_features):
            unique = np.unique(data[:, i])
            unique = unique[~np.isnan(unique)]
            
            if len(unique) < self.threshold:
                self.discrete_cols.append(i)
                # Create Grid Mapping: 0..N -> -1..1
                # e.g., 3 vals -> [-1, 0, 1]
                # e.g., 2 vals -> [-1, 1]
                n_vals = len(unique)
                if n_vals > 1:
                    grid_points = np.linspace(-1, 1, n_vals)
                else:
                    grid_points = np.array([0.0]) # Constant column
                
                self.mappings[i] = {orig: grid for orig, grid in zip(unique, grid_points)}
                # Store sorted unique vals for snapping later
                self.reverse_mappings[i] = unique
                
                if i in [25, 28]: # Confirming the problem features
                    print(f"  -> F{i+1} Locked to Grid: {unique}")
            else:
                self.continuous_cols.append(i)
                
        print(f"  Found {len(self.discrete_cols)} Discrete and {len(self.continuous_cols)} Continuous features.")
        
        # 2. Fit Continuous Scaler
        if self.continuous_cols:
            self.q_scaler.fit(data[:, self.continuous_cols])
            
    def transform(self, data):
        data_out = data.copy()
        
        # Continuous -> Quantile
        if self.continuous_cols:
            data_out[:, self.continuous_cols] = self.q_scaler.transform(data[:, self.continuous_cols])
            
        # Discrete -> Grid Lookup
        for i in self.discrete_cols:
            col_data = data[:, i]
            # Use map (slow but safe) or searchsorted (fast)
            # Since exact matches might be tricky with floats, we use searchsorted on the ORIGINAL unique values
            # But wait, input data is floats. Let's use nearest match to key.
            unique_vals = self.reverse_mappings[i]
            # Find index of nearest unique value
            diff = np.abs(col_data[:, None] - unique_vals[None, :])
            indices = diff.argmin(axis=1)
            # Map those indices to -1..1 grid
            n_vals = len(unique_vals)
            if n_vals > 1:
                grid_points = np.linspace(-1, 1, n_vals)
                data_out[:, i] = grid_points[indices]
            else:
                data_out[:, i] = 0.0
                
        return data_out
        
    def inverse_transform(self, data):
        data_out = data.copy()
        
        # Continuous -> Inverse Quantile
        if self.continuous_cols:
            data_out[:, self.continuous_cols] = self.q_scaler.inverse_transform(data[:, self.continuous_cols])
            
        # Discrete -> Snap to Grid -> Map back
        for i in self.discrete_cols:
            col_data = data[:, i]
            unique_vals = self.reverse_mappings[i]
            n_vals = len(unique_vals)
            
            if n_vals > 1:
                grid_points = np.linspace(-1, 1, n_vals)
                # Snap predicted value to nearest Grid Point
                diff = np.abs(col_data[:, None] - grid_points[None, :])
                indices = diff.argmin(axis=1)
                # Map back to original value
                data_out[:, i] = unique_vals[indices]
            else:
                data_out[:, i] = unique_vals[0]
                
        return data_out

# ==========================================
# 2. PLOTTING
# ==========================================
def plot_full_evaluation(predictions, ground_truth, imputed_input, feature_names):
    real_mask = (imputed_input != -1).astype(bool)
    missing_mask = ~real_mask 
    n_features = predictions.shape[1]
    chunk_size = 16
    
    for start_idx in range(0, n_features, chunk_size):
        end_idx = min(start_idx + chunk_size, n_features)
        current_features = range(start_idx, end_idx)
        n_current = len(current_features)
        
        n_cols = 4
        n_rows = (n_current + n_cols - 1) // n_cols
        
        # Scatter Plots
        fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 5 * n_rows))
        fig.suptitle(f"Accuracy (F{start_idx+1}-F{end_idx})", fontsize=16, y=1.02)
        axes = axes.flatten()
        
        for i, feat_idx in enumerate(current_features):
            idx = missing_mask[:, feat_idx]
            ax = axes[i]
            if idx.sum() < 2:
                ax.text(0.5, 0.5, "N/A")
                continue
            gt = ground_truth[idx, feat_idx]
            pred = predictions[idx, feat_idx]
            
            ax.scatter(gt, pred, alpha=0.3, s=10)
            vmin, vmax = min(gt.min(), pred.min()), max(gt.max(), pred.max())
            ax.plot([vmin, vmax], [vmin, vmax], 'r--')
            
            mse = mean_squared_error(gt, pred)
            r2 = r2_score(gt, pred)
            ax.set_title(f"{feature_names[feat_idx]}\nR²: {r2:.3f} | MSE: {mse:.4f}")
            
        plt.tight_layout()
        plt.show()
        
        # Distribution Plots
        fig, axes = plt.subplots(n_rows, n_cols * 2, figsize=(20, 4 * n_rows))
        fig.suptitle(f"Distributions (F{start_idx+1}-F{end_idx})", fontsize=16, y=1.02)
        axes = axes.flatten()
        
        for i, feat_idx in enumerate(current_features):
            idx = missing_mask[:, feat_idx]
            ax_gt = axes[2*i]
            ax_pred = axes[2*i+1]
            
            if idx.sum() == 0: continue
            gt = ground_truth[idx, feat_idx]
            pred = predictions[idx, feat_idx]
            
            all_d = np.concatenate([gt, pred])
            vmin, vmax = all_d.min(), all_d.max()
            
            ax_gt.hist(gt, bins=20, density=True, color='skyblue', alpha=0.8)
            ax_gt.set_title(f"{feature_names[feat_idx]} GT")
            ax_gt.set_xlim(vmin, vmax)
            
            ax_pred.hist(pred, bins=20, density=True, color='salmon', alpha=0.8)
            ax_pred.set_title(f"{feature_names[feat_idx]} Model")
            ax_pred.set_xlim(vmin, vmax)
        
        plt.tight_layout()
        plt.show()

# ==========================================
# 3. MODEL (V11 Architecture)
# ==========================================
class SinusoidalPositionEmbeddings(nn.Module):
    def __init__(self, dim):
        super().__init__()
        self.dim = dim
    def forward(self, time):
        device = time.device
        half_dim = self.dim // 2
        embeddings = math.log(10000) / (half_dim - 1)
        embeddings = torch.exp(torch.arange(half_dim, device=device) * -embeddings)
        embeddings = time[:, None] * embeddings[None, :]
        embeddings = torch.cat((embeddings.sin(), embeddings.cos()), dim=-1)
        return embeddings

class LargeDiffusion(nn.Module):
    def __init__(self, input_dim, hidden_dims=[1024, 1024, 1024, 1024]):
        super().__init__()
        self.input_dim = input_dim
        time_dim = 64
        self.time_mlp = nn.Sequential(
            SinusoidalPositionEmbeddings(time_dim),
            nn.Linear(time_dim, time_dim),
            nn.SiLU(),
            nn.Linear(time_dim, time_dim)
        )
        self.layers = nn.ModuleList()
        in_dim = (input_dim * 2) + time_dim
        for h_dim in hidden_dims:
            self.layers.append(nn.Sequential(
                nn.Linear(in_dim, h_dim),
                nn.BatchNorm1d(h_dim),
                nn.SiLU(),
                nn.Dropout(0.2)
            ))
            in_dim = h_dim
        self.final = nn.Linear(in_dim, input_dim)

    def forward(self, x, t, x_self_cond=None):
        if x_self_cond is None: x_self_cond = torch.zeros_like(x)
        t_emb = self.time_mlp(t)
        x = torch.cat([x, x_self_cond, t_emb], dim=1)
        for layer in self.layers: x = layer(x)
        return self.final(x)

class DiffusionManager:
    def __init__(self, model, device, timesteps=300):
        self.model = model
        self.device = device
        self.timesteps = timesteps
        beta_start = 0.0001
        beta_end = 0.02
        self.betas = torch.linspace(beta_start, beta_end, timesteps).to(device)
        self.alphas = 1. - self.betas
        self.alphas_cumprod = torch.cumprod(self.alphas, dim=0)
        self.sqrt_alphas_cumprod = torch.sqrt(self.alphas_cumprod)
        self.sqrt_one_minus_alphas_cumprod = torch.sqrt(1. - self.alphas_cumprod)

    def add_noise(self, x_0, t):
        noise = torch.randn_like(x_0)
        sqrt_alpha = self.sqrt_alphas_cumprod[t][:, None]
        sqrt_one_minus = self.sqrt_one_minus_alphas_cumprod[t][:, None]
        return sqrt_alpha * x_0 + sqrt_one_minus * noise, noise

    def train_step(self, batch_data, optimizer):
        self.model.train()
        optimizer.zero_grad()
        t = torch.randint(0, self.timesteps, (batch_data.shape[0],), device=self.device).long()
        x_t, noise = self.add_noise(batch_data, t)
        
        with torch.no_grad():
            if np.random.random() > 0.5:
                predicted_noise = self.model(x_t, t, x_self_cond=None)
                x_start_est = self.predict_start_from_noise(x_t, t, predicted_noise)
                x_self_cond = x_start_est.detach()
            else:
                x_self_cond = torch.zeros_like(batch_data)
                
        predicted_noise = self.model(x_t, t, x_self_cond=x_self_cond)
        loss = F.mse_loss(predicted_noise, noise)
        loss.backward()
        optimizer.step()
        return loss.item()
    
    def predict_start_from_noise(self, x_t, t, noise):
        sqrt_alpha = self.sqrt_alphas_cumprod[t][:, None]
        sqrt_one_minus = self.sqrt_one_minus_alphas_cumprod[t][:, None]
        return (x_t - sqrt_one_minus * noise) / sqrt_alpha

    @torch.no_grad()
    def impute(self, x_masked, mask, n_samples_per_instance=1):
        self.model.eval()
        n_samples, n_features = x_masked.shape
        x_masked = x_masked.repeat_interleave(n_samples_per_instance, dim=0)
        mask = mask.repeat_interleave(n_samples_per_instance, dim=0)
        x = torch.randn_like(x_masked)
        x_self_cond = torch.zeros_like(x)
        
        for i in reversed(range(self.timesteps)):
            t = torch.full((x.shape[0],), i, device=self.device, dtype=torch.long)
            predicted_noise = self.model(x, t, x_self_cond=x_self_cond)
            x_self_cond = self.predict_start_from_noise(x, t, predicted_noise)
            
            beta_t = self.betas[i]
            alpha_t = self.alphas[i]
            alpha_bar_t = self.alphas_cumprod[i]
            mean = (1 / torch.sqrt(alpha_t)) * (x - (beta_t / torch.sqrt(1 - alpha_bar_t)) * predicted_noise)
            
            if i > 0:
                noise = torch.randn_like(x)
                sigma_t = torch.sqrt(beta_t)
                x = mean + sigma_t * noise
            else:
                x = mean

            if i > 0:
                noise = torch.randn_like(x_masked)
                sqrt_alpha_prev = self.sqrt_alphas_cumprod[i-1]
                sqrt_one_prev = self.sqrt_one_minus_alphas_cumprod[i-1]
                known_noisy = sqrt_alpha_prev * x_masked + sqrt_one_prev * noise
                x = x * (1 - mask) + known_noisy * mask
            else:
                x = x * (1 - mask) + x_masked * mask
                
        return x.reshape(n_samples, n_samples_per_instance, n_features)

# ==========================================
# 4. EXECUTION
# ==========================================
splits = load_dataset_splits('dataset')
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

train_original = splits['train']['original']
GLOBAL_MIN = train_original.min(axis=0)
GLOBAL_MAX = train_original.max(axis=0)

# 1. HYBRID PREPROCESSING (V12)
# Detects 0, 0.5, 1 columns and locks them to -1, 0, 1 grid
preprocessor = GridPreprocessor(discrete_threshold=25)
preprocessor.fit(train_original)
train_data_norm = preprocessor.transform(train_original)

train_loader = DataLoader(torch.FloatTensor(train_data_norm), batch_size=64, shuffle=True)

# 2. TRAINING
model = LargeDiffusion(train_original.shape[1]).to(device)
manager = DiffusionManager(model, device, timesteps=300)
optimizer = optim.AdamW(model.parameters(), lr=2e-4, weight_decay=1e-5)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=20, factor=0.5)

print("Starting V12 (Grid Locked) Training...")
epochs = 250
train_losses = []
for epoch in range(epochs):
    epoch_loss = 0
    for batch in train_loader:
        batch = batch.to(device)
        loss = manager.train_step(batch, optimizer)
        epoch_loss += loss
    avg_loss = epoch_loss / len(train_loader)
    train_losses.append(avg_loss)
    if (epoch + 1) % 10 == 0:
        print(f"Epoch {epoch+1}/{epochs} | Loss: {avg_loss:.5f}")
        scheduler.step(avg_loss)
plt.plot(train_losses)
plt.show()

# 3. GENERATION
def generate_samples_diffusion(x_imputed, n_samples_per_test=100):
    mask = (x_imputed != -1).astype(float)
    x_clean = x_imputed.copy()
    for i in range(x_clean.shape[1]):
        col = x_clean[:, i]
        valid = col[col != -1]
        if len(valid) > 0: x_clean[col == -1, i] = np.median(valid)
        else: x_clean[col == -1, i] = 0
             
    x_norm = preprocessor.transform(x_clean) # Uses Grid + Quantile
    x_tensor = torch.FloatTensor(x_norm).to(device)
    mask_tensor = torch.FloatTensor(mask).to(device)
    
    print(f"Generating {n_samples_per_test} samples...")
    batch_size = 25
    all_outputs = []
    
    for i in range(0, len(x_tensor), batch_size):
        b_x = x_tensor[i:i+batch_size]
        b_m = mask_tensor[i:i+batch_size]
        with torch.no_grad():
            generated = manager.impute(b_x, b_m, n_samples_per_instance=n_samples_per_test)
        
        gen_np = generated.cpu().numpy()
        denorm_batch = np.zeros_like(gen_np)
        for s in range(n_samples_per_test):
            # Latent Clip
            sample_s_latent = np.clip(gen_np[:, s, :], -4.0, 4.0)
            # Inverse Transform (Handles the hard snapping)
            sample_s = preprocessor.inverse_transform(sample_s_latent)
            # Safety Clamp
            sample_s = np.clip(sample_s, GLOBAL_MIN, GLOBAL_MAX)
            
            denorm_batch[:, s, :] = sample_s
        all_outputs.append(denorm_batch)
    return np.concatenate(all_outputs, axis=0)

# 4. EVAL
print("\nINTERNAL EVALUATION")
test_imputed = splits['test']['imputed']
test_original = splits['test']['original']
eval_samples = generate_samples_diffusion(test_imputed, n_samples_per_test=10)
test_prediction_mean = eval_samples.mean(axis=1)
plot_full_evaluation(test_prediction_mean, test_original, test_imputed, [f'F{i+1}' for i in range(train_original.shape[1])])

# 5. SUBMISSION
print("\nGenerating Submission...")
X_test2 = splits['test2']['imputed']
test2_samples = generate_samples_diffusion(X_test2, n_samples_per_test=100)
np.save('submission_diffusion_v12.npy', test2_samples)
print("Saved submission_diffusion_v12.npy")
```
