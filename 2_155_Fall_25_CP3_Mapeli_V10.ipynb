You are given thousands of anonymized and normalised numeric feature vectors representing real car designs.  

However, the team remembers that the features originally came from categories like:

- **Physical geometric parameters**  
  Length, ramp angles, bumper curvature, roof curvature, panel slopes, hood angle, etc.  
  *(But you won’t know which feature corresponds to which.)*

- **Aerodynamic coefficients**  
  Drag coefficient (Cd), lift/downforce (Cl), and other flow-derived metrics.

- **Cabin and packaging descriptors**  
  Approximate cabin volume, frontal area, interior shape metrics.

Your model must learn correlations between them to generate valid completions.

To simulate real engineering constraints, **some features are revealed** (the known physics/performance requirements) and others are **masked**.  
Your AI Copilot must generate **many plausible completions** for these masked (free) parameters.

### Data Preprocessing and Missing Value Handling

This is a critical section where we prepare our data for the VAE model. Key points:

**Missing Value Conventions:**
- In CSV files: `-1` indicates missing values
- In mask files: `True` = missing, `False` = observed
- For PyTorch models: We convert to `1` = observed, `0` = missing (standard convention)

**Why This Matters:**
Our VAE needs to distinguish between observed values (which provide constraints) and missing values (which need to be generated). The mask tells the model which values to trust and which to predict.

```python
# Correcting skeweed distributions
class FeatureTransformer:
    """
    Handles preprocessing for car design features, specifically applying the 
    Yeo-Johnson transformation to skewed features (28-34) to make them more Gaussian.
    
    The transformer is fitted only on non-missing data to ensure robustness.
    """
    def __init__(self, feature_indices_to_transform: List[int]):
        """
        Initializes the transformer.
        :param feature_indices_to_transform: A list of 0-indexed column indices 
                                             corresponding to the skewed features.
        """
        self.indices = feature_indices_to_transform
        # Use a list of transformers, one for each feature
        self.transformers: List[PowerTransformer] = [
            PowerTransformer(method='yeo-johnson', standardize=True) 
            for _ in feature_indices_to_transform
        ]
        print(f"Transformer initialized for {len(self.indices)} features.")

    def fit(self, X_train_NaN: pd.DataFrame):
        """
        Fits the PowerTransformer to the non-missing training data.
        
        :param X_train_NaN: The training data as a DataFrame, with missing values 
                            represented by NaN.
        """
        if not self.indices:
            print("No features configured for transformation.")
            return

        print(f"Fitting PowerTransformer on non-missing data for indices: {self.indices}...")
        
        # 1. Select the columns that need transformation
        cols_to_transform = X_train_NaN.iloc[:, self.indices]
        
        # 2. Iterate through each feature and fit its dedicated transformer
        for i, col_index in enumerate(self.indices):
            feature_data = cols_to_transform.iloc[:, i].dropna().values.reshape(-1, 1)
            
            if len(feature_data) > 0:
                self.transformers[i].fit(feature_data)
                
        print("✓ PowerTransformer fitted successfully.")

    def transform(self, X_NaN: Union[pd.DataFrame, np.ndarray]) -> np.ndarray:
        """
        Applies the fitted transformation to the input data.
        
        :param X_NaN: The data (train, validation, or test) with NaN for missing values.
        :return: A NumPy array with the specified features transformed.
        """
        if not self.indices:
            return X_NaN.values if isinstance(X_NaN, pd.DataFrame) else X_NaN
            
        X_out = X_NaN.values.copy() if isinstance(X_NaN, pd.DataFrame) else X_NaN.copy()
        
        print(f"Transforming data for indices: {self.indices}...")
        
        for i, col_index in enumerate(self.indices):
            # Extract the column data (including NaNs)
            col_data = X_out[:, col_index].reshape(-1, 1)
            
            # Identify where the NaNs are
            nan_mask = np.isnan(col_data).flatten()
            
            # Only transform non-NaN values
            if np.sum(~nan_mask) > 0:
                # Apply the transformation
                transformed_values = self.transformers[i].transform(col_data[~nan_mask])
                
                # Place transformed values back into the output array
                X_out[~nan_mask, col_index] = transformed_values.flatten()

        return X_out
    
```

```python
# Features 28-34 correspond to 0-indexed positions 27 through 33
SKEWED_INDICES = list(range(27, 34)) 

X_train_original_c = X_train_original.copy()

df_X_train_original_c = pd.DataFrame(X_train_original_c)

# 2. Initialize and fit the transformer
preprocessor = FeatureTransformer(SKEWED_INDICES)
preprocessor.fit(df_X_train_original_c)
    
# 3. Apply the transformation
transformed_data_np = preprocessor.transform(df_X_train_original_c)
transformed_df = pd.DataFrame(transformed_data_np)
    
# 4. Check the results (Should be close to zero mean, close to zero median)
print("\n--- Transformed Skewness Check (Feature 28, index 27) ---")
print(f"Transformed Mean: {transformed_df.iloc[:, 27].mean():.3f}")
print(f"Transformed Median: {transformed_df.iloc[:, 27].median():.3f}")
    
# Note: NaN values remain NaN after transformation
print(f"NaN count after transform: {np.isnan(transformed_data_np[:, 27]).sum()}")
```

```python
import numpy as np
import pandas as pd
import lightgbm as lgb
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from tqdm import tqdm
import warnings

# Import your utils
from utils import load_dataset_splits

warnings.filterwarnings('ignore')

# ==========================================
# 1. ITERATIVE GBDT IMPUTER (The Engine)
# ==========================================
class GBDTImputer:
    def __init__(self, discrete_cols, n_cycles=1):
        self.discrete_cols = discrete_cols # List of indices
        self.n_cycles = n_cycles
        self.models = {} # {feature_idx: model}
        self.residuals = {} # {feature_idx: std_dev_of_residuals}
        
    def _get_valid_cols(self, data, target_col):
        # Use all columns except target as predictors
        return [i for i in range(data.shape[1]) if i != target_col]

    def fit(self, data, mask):
        """
        Trains a model for every feature that has missing values.
        data: (N, F) - Should be initialized (e.g., with mean/median)
        mask: (N, F) - 1=Observed, 0=Missing
        """
        n_samples, n_features = data.shape
        print(f"Training LightGBM Imputers for {n_features} features...")
        
        # We perform a few cycles to refine the "ground" (input features)
        for cycle in range(self.n_cycles):
            print(f"  Cycle {cycle+1}/{self.n_cycles}...")
            
            for target_col in range(n_features):
                # 1. Define Training Data
                # We only train on OBSERVED rows for this feature
                observed_idx = np.where(mask[:, target_col] == 1)[0]
                
                if len(observed_idx) == 0:
                    continue # Skip if completely missing
                    
                X_train = data[observed_idx][:, self._get_valid_cols(data, target_col)]
                y_train = data[observed_idx, target_col]
                
                # 2. Select Model Type
                if target_col in self.discrete_cols:
                    # CLASSIFICATION for discrete columns
                    # Need to map floats (0.0, 0.5, 1.0) to integers (0, 1, 2) for LGBM
                    unique_vals = np.unique(y_train)
                    label_map = {val: i for i, val in enumerate(unique_vals)}
                    y_train_mapped = np.array([label_map[val] for val in y_train])
                    
                    model = lgb.LGBMClassifier(
                        verbosity=-1, 
                        n_estimators=100,
                        random_state=42
                    )
                    model.fit(X_train, y_train_mapped)
                    
                    # Store mapping to reverse later
                    model.label_map_ = label_map
                    model.inverse_map_ = {v: k for k, v in label_map.items()}
                    
                    # For classification, "residuals" isn't quite right, but we skip noise for discrete
                    # We rely on probability sampling for diversity
                    self.residuals[target_col] = None
                    
                else:
                    # REGRESSION for continuous
                    model = lgb.LGBMRegressor(
                        verbosity=-1,
                        n_estimators=100,
                        random_state=42
                    )
                    model.fit(X_train, y_train)
                    
                    # Calculate residuals (RMSE) to simulate noise later
                    preds = model.predict(X_train)
                    rmse = np.sqrt(mean_squared_error(y_train, preds))
                    self.residuals[target_col] = rmse
                
                self.models[target_col] = model
                
                # 3. UPDATE DATA IMMEDIATELY (Iterative Refinement)
                # Predict missing values for this column to improve inputs for next feature
                missing_idx = np.where(mask[:, target_col] == 0)[0]
                if len(missing_idx) > 0:
                    X_miss = data[missing_idx][:, self._get_valid_cols(data, target_col)]
                    if target_col in self.discrete_cols:
                        mapped_preds = model.predict(X_miss)
                        preds = np.array([model.inverse_map_[p] for p in mapped_preds])
                    else:
                        preds = model.predict(X_miss)
                    
                    # Update matrix
                    data[missing_idx, target_col] = preds

    def impute_stochastic(self, data, mask, n_samples_per_test=100):
        """
        Generates multiple samples.
        Continuous: Prediction + GaussianNoise(0, residual_std)
        Discrete: Weighted sampling from predicted probabilities.
        """
        n_samples, n_features = data.shape
        # Output shape: (N, 100, F)
        final_samples = np.zeros((n_samples, n_samples_per_test, n_features))
        
        # Initialize with current state
        # We repeat the data structure 100 times
        print(f"Generating {n_samples_per_test} stochastic variations...")
        
        for col in range(n_features):
            missing_idx = np.where(mask[:, col] == 0)[0]
            if len(missing_idx) == 0:
                # If no missing, just copy observed
                final_samples[:, :, col] = data[:, col][:, None]
                continue
                
            # Prepare Input (Using current best estimate from 'data')
            X_miss = data[missing_idx][:, self._get_valid_cols(data, col)]
            model = self.models.get(col)
            
            if model is None: 
                # Fallback
                final_samples[:, :, col] = data[:, col][:, None]
                continue
                
            if col in self.discrete_cols:
                # --- DISCRETE: PROBABILITY SAMPLING ---
                # Get probs [0.1, 0.8, 0.1]
                probs = model.predict_proba(X_miss) # (N_missing, N_classes)
                classes = list(model.inverse_map_.values()) # e.g. [0.0, 0.5, 1.0]
                
                # Sample for each of the 100 versions
                for s in range(n_samples_per_test):
                    # Vectorized sampling
                    # We use random uniform vs cumulative sum
                    cum_probs = probs.cumsum(axis=1)
                    r = np.random.rand(len(probs), 1)
                    # argargmax(cum_probs > r) gives first index where cumsum > random
                    # This selects class based on weight
                    choice_indices = (cum_probs < r).sum(axis=1)
                    
                    # Map back to values
                    choices = np.array([classes[i] for i in choice_indices])
                    
                    # Assign
                    final_samples[missing_idx, s, col] = choices
                    
            else:
                # --- CONTINUOUS: MEAN + NOISE ---
                base_preds = model.predict(X_miss)
                sigma = self.residuals[col]
                
                # Standard noise scaling factor (tune this if variance is too high/low)
                # If sigma is 0.1, noise will be +/- 0.1 approx
                noise_scale = 1.0 
                
                for s in range(n_samples_per_test):
                    noise = np.random.normal(0, sigma * noise_scale, size=len(base_preds))
                    final_samples[missing_idx, s, col] = base_preds + noise
            
            # For observed rows, copy original
            observed_idx = np.where(mask[:, col] == 1)[0]
            final_samples[observed_idx, :, col] = data[observed_idx, col][:, None]
            
        return final_samples

# ==========================================
# 2. PLOTTING
# ==========================================
def plot_full_evaluation(predictions, ground_truth, imputed_input, feature_names):
    real_mask = (imputed_input != -1).astype(bool)
    missing_mask = ~real_mask 
    n_features = predictions.shape[1]
    chunk_size = 16
    
    for start_idx in range(0, n_features, chunk_size):
        end_idx = min(start_idx + chunk_size, n_features)
        current_features = range(start_idx, end_idx)
        n_current = len(current_features)
        
        n_cols = 4
        n_rows = (n_current + n_cols - 1) // n_cols
        
        # Scatter Plots
        fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 5 * n_rows))
        fig.suptitle(f"Accuracy (F{start_idx+1}-F{end_idx})", fontsize=16, y=1.02)
        axes = axes.flatten()
        
        for i, feat_idx in enumerate(current_features):
            idx = missing_mask[:, feat_idx]
            ax = axes[i]
            if idx.sum() < 2:
                ax.text(0.5, 0.5, "N/A")
                continue
            gt = ground_truth[idx, feat_idx]
            pred = predictions[idx, feat_idx]
            
            ax.scatter(gt, pred, alpha=0.3, s=10)
            vmin, vmax = min(gt.min(), pred.min()), max(gt.max(), pred.max())
            ax.plot([vmin, vmax], [vmin, vmax], 'r--')
            
            mse = mean_squared_error(gt, pred)
            r2 = r2_score(gt, pred)
            ax.set_title(f"{feature_names[feat_idx]}\nR²: {r2:.3f} | MSE: {mse:.4f}")
            
        plt.tight_layout()
        plt.show()
        
        # Distribution Plots
        fig, axes = plt.subplots(n_rows, n_cols * 2, figsize=(20, 4 * n_rows))
        fig.suptitle(f"Distributions (F{start_idx+1}-F{end_idx})", fontsize=16, y=1.02)
        axes = axes.flatten()
        
        for i, feat_idx in enumerate(current_features):
            idx = missing_mask[:, feat_idx]
            ax_gt = axes[2*i]
            ax_pred = axes[2*i+1]
            
            if idx.sum() == 0: continue
            gt = ground_truth[idx, feat_idx]
            pred = predictions[idx, feat_idx]
            
            all_d = np.concatenate([gt, pred])
            vmin, vmax = all_d.min(), all_d.max()
            
            ax_gt.hist(gt, bins=20, density=True, color='skyblue', alpha=0.8)
            ax_gt.set_title(f"{feature_names[feat_idx]} GT")
            ax_gt.set_xlim(vmin, vmax)
            
            ax_pred.hist(pred, bins=20, density=True, color='salmon', alpha=0.8)
            ax_pred.set_title(f"{feature_names[feat_idx]} Model")
            ax_pred.set_xlim(vmin, vmax)
        
        plt.tight_layout()
        plt.show()

# ==========================================
# 3. EXECUTION
# ==========================================
splits = load_dataset_splits('dataset')

# 1. DATA PREP
# We need a clean initialization. Mean filling is standard for IterativeImputer init.
train_original = splits['train']['original']
val_original = splits['val']['original']

# Combine Train+Val for maximum learning
# (Normally we split, but for imputation competition, more data = better mapping)
X_full = np.vstack([train_original, val_original])
# Create mask (1=Observed) - Here everything is observed
mask_full = np.ones_like(X_full)

# 2. IDENTIFY DISCRETE COLUMNS
# You said F25, F26, F29 are 0,0.5,1. F12, F15 are also discrete.
# We'll scan like before.
print("Analyzing features...")
discrete_cols = []
for i in range(X_full.shape[1]):
    unique = np.unique(X_full[:, i])
    if len(unique) < 25:
        discrete_cols.append(i)
print(f"Detected Discrete Columns (Will use Classifier): {discrete_cols}")

# 3. TRAIN IMPUTER
# We train on the COMPLETE data to learn the relationships
imputer = GBDTImputer(discrete_cols=discrete_cols, n_cycles=2)
imputer.fit(X_full, mask_full)

# 4. EVALUATION (On Test Split)
print("\nINTERNAL EVALUATION")
test_imputed = splits['test']['imputed'] # Has -1s
test_original = splits['test']['original'] # Ground Truth
test_mask = (test_imputed != -1).astype(int) # 1=Observed

# Replace -1 with mean for initial pass (LGBM handles it, but we need valid matrix)
# Actually, we can pass -1 if we handle it in `_get_valid_cols`. 
# But for simplicity, let's simple-impute the input first.
X_test_init = test_imputed.copy()
for i in range(X_test_init.shape[1]):
    col = X_test_init[:, i]
    valid = col[col != -1]
    mean_val = np.mean(valid) if len(valid) > 0 else 0
    X_test_init[col == -1, i] = mean_val

# Generate 10 samples for eval
eval_samples = imputer.impute_stochastic(X_test_init, test_mask, n_samples_per_test=10)
test_prediction_mean = eval_samples.mean(axis=1)

# Score
mse_scores = []
input_dim = test_original.shape[1]
for i in range(input_dim):
    missing_idx = np.where(test_mask[:, i] == 0)[0]
    if len(missing_idx) > 0:
        mse = mean_squared_error(test_original[missing_idx, i], test_prediction_mean[missing_idx, i])
        mse_scores.append(mse)

print(f"\nAverage MSE on Missing Values: {np.mean(mse_scores):.4f}")
plot_full_evaluation(test_prediction_mean, test_original, test_imputed, [f'F{i+1}' for i in range(input_dim)])

# 5. SUBMISSION
print("\nGenerating Submission...")
X_test2 = splits['test2']['imputed']
mask_test2 = (X_test2 != -1).astype(int)

# Init fill
X_test2_init = X_test2.copy()
for i in range(X_test2_init.shape[1]):
    col = X_test2_init[:, i]
    valid = col[col != -1]
    mean_val = np.mean(valid) if len(valid) > 0 else 0
    X_test2_init[col == -1, i] = mean_val

test2_samples = imputer.impute_stochastic(X_test2_init, mask_test2, n_samples_per_test=100)
np.save('submission_xgboost_v13.npy', test2_samples)
print("Saved submission_xgboost_v13.npy")
```
